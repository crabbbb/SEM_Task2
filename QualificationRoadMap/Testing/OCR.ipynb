```python
import requests
from bs4 import BeautifulSoup

# Example of fetching free proxy list from a website
url = 'https://free-proxy-list.net/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Parse the proxy list (this example depends on the website's structure)
proxy_table = soup.find('table')

# Collect all proxies (IP:Port) into a list
proxies = []
for row in proxy_table.tbody.find_all('tr'):
    cells = row.find_all('td')
    ip = cells[0].text
    port = cells[1].text
    proxy = f"http://{ip}:{port}"
    proxies.append(proxy)

print(proxies)
# print(proxy_table)
```

```python
import requests
import random
from urllib.parse import quote

# Ensure the URL is properly encoded
url_to_request = 'http://example.com'


# Make a request with the properly encoded URL
# response = requests.get(safe_url, proxies=proxies)

# Function to fetch proxies from ProxyScrape API
def get_proxies():
    url = "https://www.proxyscrape.com/api/v1/?request=getproxies&proxytype=http&timeout=1000&country=all&ssl=yes&anonymity=elite"
    response = requests.get(url)
    proxy_list = response.text.split('\n')
    proxy_list = [proxy.strip() for proxy in proxy_list if proxy.strip()]  # Clean and filter empty lines
    return proxy_list

# Function to make request using a proxy
def make_request_with_proxy(url, proxy):
    proxies = {
        'http': f'http://{proxy}',
        'https': f'http://{proxy}',  # Sometimes HTTPS proxies may not work reliably for HTTP requests
    }
    try:
        safe_url = quote(url, safe=":/")
        response = requests.get(safe_url, proxies=proxies, timeout=5)
        print(f"Proxy Used: {proxy}")
        print(f"Status Code: {response.status_code}")
        print(response.text[:500])  # Print first 500 characters of the response
    except requests.exceptions.ProxyError:
        print(f"Proxy Error with: {proxy}")
    except requests.exceptions.Timeout:
        print(f"Timeout Error with: {proxy}")
    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")

# Main function
def main():
    url_to_request = 'http://example.com'  # Replace with the target URL
    proxy_list = get_proxies()  # Get proxy list from ProxyScrape
    
    if proxy_list:
        proxy = random.choice(proxy_list)  # Pick a random proxy from the list
        make_request_with_proxy(url_to_request, proxy)
    else:
        print("No proxies found.")

# Run the main function
if __name__ == "__main__":
    main()
```
